```
import numpy as np
import math 
import matplotlib.pyplot as plt



#classe de regressao simples
class regressao:
    def __init__(self):
        pass

    def fit(self, amostra, rotolos):

        #tirando a media x = amostra, y = rotolos (ou observacao)
        media_x = np.mean(amostra)
        media_y = np.mean(rotolos)


        #Calculo do Beta1
        self.beta1 = np.sum((amostra - media_x)*(rotolos - media_y)) / np.sum((amostra - media_x)**2)
        #Calculo do Beta0
        self.beta0 = media_y - self.beta1*media_x

    def predict(self, amostra):

        # Y e o valor predito calculado
        self.Y = self.beta0 + self.beta1*amostra

        return self.Y

    def RSS(self, rotolos):
    	RSS = np.sum((rotolos - self.Y)**2)
    	return RSS

    def MSE(self, rotolos):
      self.n = rotolos.shape[0]
      MSE = self.RSS(rotolos)/self.n
      return MSE

    def R2(self, rotolos):
        TSS = np.sum((rotolos - np.mean(rotolos))**2)

        R2 = 1 - self.RSS(rotolos)/TSS

        return R2



data = np.loadtxt("Dados/housing.data", delimiter=",")

np.random.shuffle(data) #embaralha valores do data

x = data[:,12]
double_x = x * x
triple_x = double_x * x
y = data[:,13]

percentual = int(x.shape[0]*0.8) # 80% dos valores de amostra de x
x_train	= x[:percentual] #pegando 80% dos valores de amostra de x (FIT)
y_train	= y[:percentual] #pegando 80% dos valores de amostra de y (FIT)
x_teste = x[percentual:] #pegando 20% dos valores de amostra de x (PREDICT)
y_teste = y[percentual:] #pegando 20% dos valores de amostra de y (PREDICT)

percentual_double = int(double_x.shape[0]*0.8)
x_train_double	= double_x[:percentual] #pegando 80% dos valores de amostra de x (FIT)
x_teste_double =  double_x[percentual:] #pegando 20% dos valores de amostra de x (PREDICT)

percentual_triple = int(triple_x.shape[0]*0.8)
x_train_triple	= triple_x[:percentual] #pegando 80% dos valores de amostra de x (FIT)
x_teste_triple  = triple_x[percentual:] #pegando 20% dos valores de amostra de x (PREDICT)



regressor = regressao()

#TREINANDO COM OS VALORES DE 80%
regressor.fit(x_train,y_train)     

#VALORES PREDITOS DE TRAIN
predict_train = regressor.predict(x_train)	
print("VALORES PREDITOS:")
print("VALORES MSE =", regressor.MSE(y_train))
print("VALORES R2 =",  regressor.R2(y_train))
print()

#VALORES PREDITOS DE TESTE
print("VALORES TESTE:")
predict_teste = regressor.predict(x_teste)
print("VALORES MSE =", regressor.MSE(y_teste))
print("VALORES R2 =",  regressor.R2(y_teste))
print()

#TREINANDO COM OS VALORES DE 80% DOUBLE
regressor.fit(x_train_double,y_train)     

#VALORES PREDITOS DE TRAIN
predict_train_double = regressor.predict(x_train_double)	
print("VALORES DOUBLE PREDITOS:")
print("VALORES MSE =", regressor.MSE(y_train))
print("VALORES R2 =",  regressor.R2(y_train))
print()

#VALORES PREDITOS DE TESTE
predict_teste_double = regressor.predict(x_teste_double)
print("VALORES DOUBLE TESTE:")
print("VALORES MSE =", regressor.MSE(y_teste))
print("VALORES R2 =",  regressor.R2(y_teste))
print()

#TREINANDO COM OS VALORES DE 80% TRIPLE
regressor.fit(x_train_triple,y_train)     

#VALORES PREDITOS DE TRAIN
predict_train_triple = regressor.predict(x_train_triple)	
print("VALORES TRIPLE PREDITOS:")
print("VALORES MSE =", regressor.MSE(y_train))
print("VALORES R2 =",  regressor.R2(y_train))
print()

#VALORES PREDITOS DE TESTE
predict_teste_triple = regressor.predict(x_teste_triple)
print("VALORES TRIPLE TESTE:")
print("VALORES MSE =", regressor.MSE(y_teste))
print("VALORES R2 =",  regressor.R2(y_teste))
print()

        valor_aux = np.linalg.inv(np.matmul(x.T,x))
        self.beta = np.matmul(np.matmul(valor_aux,x.T),y)
        
self.Beta =  np.matmul(np.matmul(np.linalg.inv(np.matmul(np.transpose(amostra), amostra)),np.transpose(amostra)),rotolos)
```
